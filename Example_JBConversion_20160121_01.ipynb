```python
def JB_to_list_benchmark(data_file):
    """"Converts The Jim Booth Format to list benchmark format for tables, [[Metadata list of strings],[Column Names],[
    list of Data Rows]]"""
    if type(data_file) is str:
        meta_list=[]
        column_names=[]
        data=[]
        column_name_line=0
        in_file=open(data_file)
        for line in in_file:
            if line[0] is '#':
                column_name_line+=1
                meta_list.append(line)
            else:
                data.append(line.split(' '))
        column_names=meta_list[column_name_line-1].split(' ')
    benchmark_list=[meta_list,column_names,data]
    return benchmark_list
```

```python
def JB_to_csv_lines(data_file):
    """"Converts The Jim Booth Format to csv_lines, adds a semi-colon after date and replaces spaces in column names with , """
    if type(data_file) is str:

        csv_line_list=[] #intialize the buffer
        column_name_line=0
        in_file=open(data_file)
        for line in in_file:
            if line[0] is '#':
                line=line.replace('M D','M; D').replace(',',';')
                column_name_line+=1
                csv_line_list.append(line)
            else:
                csv_line_list.append(line.replace(' ',','))
        csv_line_list[column_name_line-1]=csv_line_list[column_name_line-1].replace(' ',',').replace('#','')
    
    return csv_line_list
```

```python
def JB_to_s2pv1(input_file_name,output_file_name,freq_units="GHz"):
    """Given a JB it writes it using ! for comments and 2 spaces for deliminator, and the option line
    assumes the current units are Hz. Note JB files can be switch terms in Foward/Reverse"""
    # if there is no extension or it is not  make it .s2p
    if output_file_name.split('.')[-1] not in ["s2p","snp"]:
        output_file_name=file_name+'.s2p'
    #open the file and clean up the metadata lines
    freq_divider={"Hz":1,"KHz":10**3,"MHz":10**6,'GHz':10**9}
    in_file=open(input_file_name)
    formatted_lines=[]
    comment_line_number=0
    for line in in_file:
        if line[0] in ['#']:
            line=line.replace('#','!')
            line=line.replace('M D','M, D').replace(';',',').replace('Hz',freq_units)
            formatted_lines.append(line)
            comment_line_number+=1
        else:
            split_line=line.split(' ')
            split_line[0]="{:10f}".format(float(split_line[0])/float(freq_divider[freq_units]))
            new_line=''
            for i,item in enumerate(split_line):
                if i==0:
                    new_line=item
                else:
                    new_line=new_line+'  '+item           
            formatted_lines.append(new_line)
    formatted_lines.insert(comment_line_number,'# %s S RI R 50 \n'%freq_units)
    out_file=open(output_file_name,'w')
    for line in formatted_lines:
        out_file.write(line)
    out_file.close()

        
            
```

```python
def convert_JB_to_s2pv1_script(directory,new_directory=None):
    """converts all JB to s2pv1 format in a directory"""
    # get all the files with out extensions in the directory
    file_names=[]
    new_names=[]
    for file_name in os.listdir(directory):
        if not fnmatch.fnmatch(file_name, '*.*'):
            file_names.append(file_name)
            #print file_names
    if new_directory is None:
        for file_name in file_names:
            new_names.append(os.path.join(directory,file_name+'.s2p'))
    else:
        for file_name in file_names:
            new_names.append(os.path.join(new_directory,file_name+'.s2p'))
    for i,file_name in enumerate(file_names):
        JB_to_s2pv1(os.path.join(directory,file_name),new_names[i])
    print "Files Converted: %s"%(str(file_names))
    print "\n"
    print "Resulting Files: %s"%(str(new_names))
    print "-"*80
    print "\n"
```

```python
# if I wanted to turn it into an xml chunk without parsing the dom I could do the following
def dictionary_to_xmlchunk(dictionary, level='attribute'):
    "returns a string formatted as xml when given a python dictionary"
    if type(dictionary) is not dict:raise
    if re.match('attr',level,re.IGNORECASE):
        prefix="<Tuple "
        postfix=" />"
        inner=""
        for key,value in dictionary.iteritems():
            inner=inner+"%s='%s' "%(key,value)
        xml_out=prefix+inner+postfix
            
    if re.match('tag',level,re.IGNORECASE):
        xml_out=""
        for key,value in dictionary.iteritems():
            xml_out=xml_out+"<"+str(key)+">"+str(value)+"</"+str(key)+">"
    return xml_out
 
 
        
        
```

```python
def dictionary_list_to_xmlchunk(list_of_dictionaries,level='attribute'):
    """returns a string of xml given a list of dictionaries, level says if keys become attribute names or tag names"""
    if type(list_of_dictionaries) is not list:raise
    xml_out=""
    for item in list_of_dictionaries:
        xml_out=xml_out+dictionary_to_xmlchunk(item,level)+"\n"
    return xml_out
```
