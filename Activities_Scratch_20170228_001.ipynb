<li><h2>Data Interoperability</h2></li>
<h3>Summary</h3>
The efficient design, simulation, analysis and dissemination of data requires the use of many tools. These tools can be commercial packages, internal developed specialty software, or open source packages. These tools frequently have different capabilities and data requirements even when working on problems in the same scientific domain. In order to facilitate comparisons, data exchange and integrated work flows data must be shared between programs and people. There have been many movements by computationally based communities to address this issue through the introduction of open and broadly encompassing formats such as xml, hdf, json, net cdf, and html. While each of these formats has provided large gains in data interoperability, they neglect a single simple truth, no one format can address the data needs of all users. There is always a task for which a data format is not ideally suited. If we view the data encapsulated by any one format as a representation of content, and each potential format as just a representation of this content, there is a direct way to connect all the formats that can represent this content. By viewing each format as an equally valid expression of the content instead of creating a hierarchal scheme that emphasizes the most widely accepted format at the time, we can in practice select the format best suited for any one task, use the tool that is optimized to this task, and transform the result for optimized performance in the next step in our work flow.
<h3>Description</h3>
This project is primarily a software related project. It requires skilled technical labor (), computing hardware (), access to a large range of software (scales as the number of formats desired). To implement this expanded interoperability approach, a network graph based approach will be employed. This network graph would have as it nodes formats of interest, and as it edges functions that moved from one format to another. For the many formats that preserve content, these nodes can be made equivalent by preserving the content of interest. To be truly accessible, it is also desirable to have the software available as a web application so that the full functionality can be expanded to a large range of stakeholders. It is anticipated that this approach will unlock tools that developed in CTL for use by other stakeholders.

<li><h2>Metadata for Improved Discoverability, Dissemination and Analysis</h2></li>
<h3>Summary</h3>

This type of description allows for researchers to disseminate their collected data to other humans and provide inputs for other analytic techniques. It is a primary goal to create a standardized way of providing descriptions to the acquired data, work flow documentation, experimental design, experimental setup and experimental state during data collection.
