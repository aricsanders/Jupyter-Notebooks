```python
def fake_data(data):
    """Fake data is a method decorator that returns a set of fake data if the instrument mode is
    self.fake_mode=True.
    For example just add @fake_data(data_to_return) before a an instrument method"""
    def method_decorator(method):
        def return_data(self,*args,**kwargs):
            if self.fake_mode:
                return data
            else:
                return method(self,*args,**kwargs)
        return return_data
    return method_decorator

```

```python
class VNAOverload(VNA):
    @fake_data(FAKE_S2P)
    def measure_s2p(self):
        return FAKE_S2P
```

```python
vna=VNAOverload("help")
```

```python
s2p=vna.measure_s2p()
```
