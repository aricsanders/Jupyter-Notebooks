# Comparison of Phase Calibration for LSNA with a Comb and a AWG
The calibration for phase in Large Signal Network Analysis requires two components; a relative phase reference to synchronize measurements at different frequencies and an absolute phase reference. Both of these functions are typically satisfied by a frequency comb that can produce harmonics of a single fundamental frequency (typically 10 MHz ). We are researching the possibility of exchanging both of these combs with a AWG that produces combs at only the frequencies of interest. The AWG has an external clock provided by a vector signal generator that is then passed to a calibrated oscilloscope to provide a time base correction signal. The AWG then produces a trigger signal on every repeat of the multisine used for calibration to trigger the oscilloscope. The phase calibration is then performed a tone at a time, by producing a single sine wave and then simultaneously measuring it with the calibrated Oscilloscope and the VNA. 

## Experimental Setup 
<img src="./AWG_Comb_Comparison_Files/Comb_Comparison_Setup.png"/>

## Experimental Method

1. Set the master clock frequency of 10 GHz for a single stable sine wave on the VSG
2. Create and load a Multisine of the proper number of tones and frequency offset to the AWG. This remains stable for the phase calibration process.
3. Create a single sine wave at the frequency of a tone in the reference multisine and measure it simultaneously on the oscilloscope and VNA. Repeat for each frequency of interest. The receiver value measured by the VNA must be constantly corrected to the measurement phase reference (Multisine). The oscilloscope measurement is transformed into the Fourier domain via a simple sine fit that determines phase and amplitude. The frequency of the fit is fixed to the frequency of interest. 

## Example Process to Compare Comb Method with AWG Method

### 1. Determine Number of Tones, Frequency Offset and Center Frequency of Interest.
In our case, we have 11 tones, a center frequency of 1 GHz, offsets of 10 MHz, 5 MHz, 2.5 MHz, 250 KHz, 25 KHz 
### 2. Determine Multisine Relative Phase (Zero or Schroeder)
We have tried both (No Major difference)
### 3.  Write Multisine to Channel 2 of AWG that is connected through the back panel to Port 4 of the VNA. 
Port is terminated in a short.
### 4. Start at Tone 1, by Writing a Single Sine Wave to Channel 1 of the AWG that is connected through the back panel to Port 1 of the VNA. 
This sine wave should be time synchronized with the multisine on port 4. In order to establish this we actually write the sine wave 10 times with different phases. Each phase is ~ 36 Degrees from the last.
### 5. Measure the Calibration Sine Wave with The VNA and Oscilloscope.
The measurement of the VNA is a ratio of the a1 receiver to the a4 receiver (it uses the multisine as a phase reference). The oscilloscope is normally a single frame containing a few periods (1-10) of the calibration sine wave, it is fit using a least squares sine function that has parameters of amplitude and phase. At the same time for frequencies that the combs have harmonics (2.5 MHz, 5 MHz and 10 MHz) we measure b3 (calibration comb) and b2 (phase reference comb). It should be noted that the power generated by the combs are small and the signals must be averaged.

```python
# we can now fix the phases using a simple function that undoes 90 degree flips and unwraps the phase
def fix_phase(phase_list):
    """Unwraps phase and fixes any 90 degree phase flips"""
    last_phase=phase_list[0]
    new_phase_list=[last_phase]
    for phase in phase_list[1:]:
        phase_offset=phase-last_phase
        #print("Phase Offset is {0}".format(phase_offset))
        if 90<phase_offset<270:
            new_phase=phase-180
        elif -90>phase_offset>-270:
            new_phase=phase+180
        elif -270>phase_offset:
            new_phase=phase+360
        else:
            new_phase=phase
        #print(" New Phase is {0}, Old Phase is {1}".format(new_phase,phase))
        new_phase_list.append(new_phase)
        last_phase=new_phase
    return new_phase_list
plt.plot(fix_phase(summary_table["Scope_Phase"][0:10]),'bo',label="Oscilloscope")
plt.plot(fix_phase(summary_table["Vna_Phase"][0:10]),'r^',label="Vna")
plt.title("Corrected Phases Measured for Tone 1/11")
plt.xlabel("Phase Set Number")
plt.ylabel("Phase (degrees)")
plt.legend()
plt.show()
```

```python
# Now we can see the difference is constant and gives us our phase calibration
vna_phases=np.array(fix_phase(summary_table["Vna_Phase"][0:10]))
scope_phases=np.array(fix_phase(summary_table["Scope_Phase"][0:10]))
phase_differences=vna_phases-scope_phases
plt.close()
plt.hist(phase_differences)
plt.title("Phase Calibration for Tone 1/11, Mean:{0:3.2f}, Std:{1:3.2f}, Range:{2:3.2f}".format(np.mean(phase_differences),
                                                                               np.std(phase_differences),
                                                                                                max(phase_differences)-min(phase_differences)))
plt.xlabel("Phase Difference")
plt.ylabel("Frequency")
plt.show()
```

```python
# Now we can repeat this for each tone to get the phase calibration for this tone spacing and tone number
summary_file_model=summary_table.copy()
tones=[]
mean_differences=[]
deviation_differences=[]
for tone_index in range(11):
    tones.append(tone_index+1)
    start_index=tone_index*(10)
    stop_index=tone_index*(10)+10
    x_data=summary_file_model["AWG_Set_Phase"][start_index:stop_index]
    #corrected_list=360+np.array(summary["Vna_Phase"][8:10])
    vna_data=np.array(fix_phase(summary_file_model["Vna_Phase"][start_index:stop_index]))
    #corrected_scope_list=180+np.array(summary["Scope_Phase"][5:10])
    scope_data=np.array(fix_phase(summary_file_model["Scope_Phase"][start_index:stop_index]))
    mean_differences.append(np.mean(vna_data-scope_data))
    deviation_differences.append(np.std(vna_data-scope_data))
plt.close()
plt.errorbar(tones,y=mean_differences,yerr=deviation_differences,fmt="b-o")
plt.xlabel("Tone")
plt.ylabel("Phase Offset")
plt.title(r"Phase Offset, "+"$F_0 = 1 GHz, \Delta F={0:.1e} Hz$".format(summary_file_model.metadata["tone_spacing"]))
plt.show()
# plt.close()
# plt.errorbar(tones,y=fix_phase(mean_differences),yerr=deviation_differences,fmt="b-o")
# plt.xlabel("Tone")
# plt.ylabel("Phase Offset")
# plt.title(r"Corrected Phase Offset, "+"$F_0 = 1 GHz, \Delta F={0:.1e} Hz$".format(summary_file_model.metadata["tone_spacing"]))
# plt.show()
```

### Analysis of Comb Data
The combs produce a very small intensity frequency comb at a spacing of 2.5 MHz (10 MHz / 4). If we take the instantaneous ratio of A1/B2 the value is too noisy to determine the phase, so we average and then calculate the phase properties.

```python
# At each point, tone and phase of tone, we acquire all of the receivers A1,A4,B2,B3 that have 
# A phase measurement reference or calibration reference. A4, B2, and B3 should not change between settings on A1. 
# We can use this as 110 measurements of the phase references. First we import a single sweep as a reference:
# Note there are two naming issues: 1. the units on the spacing are Hz not KHz, 2. Tone Number Begins with 0 instead of 1
single_vna_sweep_path=r"C:\Share\AWG_Phase_Calibration\Schroeder_Phase_10Hz\11_Tones_10000KHz_Spacing\Tone_1_Phase_000_Calibration_Vna_Reciever_Values_20181010_001.txt"
single_vna_sweep_schema=read_schema(change_extension(single_vna_sweep_path,new_extension="schema"))
single_vna_sweep_table=AsciiDataTable(single_vna_sweep_path,**single_vna_sweep_schema)
```

```python
# Now the information saved can be inferred through the column names
single_vna_sweep_table.column_names
```

```python
# Now we can plot the two comb amplitudes and the A4 amplitude to compare them
# We need to convert each to a complex array
phase_reference_receievers=["A4","B2","B3"]
complex_single_sweep=[]
for receiver in phase_reference_receievers:
    reciever_data=single_vna_sweep_table[["re"+receiver,"im"+receiver]]
    complex_receiver_data=map(lambda x:complex(x[0],x[1]),reciever_data)
    complex_array=np.array(complex_receiver_data)
    complex_single_sweep.append(complex_array)
x_data=10**-9*np.array(single_vna_sweep_table["Frequency"])
plt.close()
figure,axes=plt.subplots(ncols=1,nrows=2,sharex=True)
axes[0].plot(x_data,20*np.log10(abs(complex_single_sweep[0])),"o",
             label="A4 = AWG Measurement Multisine Reference")
axes[0].plot(x_data,20*np.log10(abs(complex_single_sweep[1])),"^",
             label="B2 = Measurement Reference Comb")
axes[0].plot(x_data,20*np.log10(abs(complex_single_sweep[2])),"s",
             label="B3 = Calibration Reference Comb")
axes[0].set_title("Uncalibrated Reciever Reading")
axes[0].set_ylabel("Reading (dB)")
axes[1].plot(x_data,np.angle(complex_single_sweep[0],deg=True),"o",
             label="A4 = AWG Measurement Multisine Reference")
axes[1].plot(x_data,np.angle(complex_single_sweep[1],deg=True),"^",
             label="B2 = Measurement Reference Comb")
axes[1].plot(x_data,np.angle(complex_single_sweep[2],deg=True),"s",
             label="B3 = Calibration Reference Comb")
axes[1].set_xlabel("Frequency (GHz)")
axes[1].set_ylabel("Phase (Degrees)")
axes[0].legend(loc="center left", bbox_to_anchor=(1.05, .5), 
                          title="Single Sweep", fancybox=True)
plt.tight_layout()
plt.show()             
```

## Generating .s1p of Short, Open Load

```python
# Extracting them from the receiver values, do this tone by tone
top_directory=r"C:\Share\AWG_Phase_Calibration\Zero_Phase_10Hz_Short"
for data_directory in os.listdir(top_directory):
    name=data_directory
    data_directory=os.path.join(top_directory,data_directory)
    file_names=os.listdir(data_directory)
    tones=[i for i in range(11)]
    s1p_complex_data=[]
    for tone in tones:
        analysis_file_names=[]
        for file_name in file_names:
            if re.search("txt",file_name,re.IGNORECASE):
                if re.search("rec",file_name,re.IGNORECASE):
                    if re.search("Tone_{0}_P".format(tone),file_name,re.IGNORECASE):
                        analysis_file_names.append(file_name)
        # Then we load all the single sweeps
        analysis_file_names=map(lambda x:os.path.join(data_directory,x),analysis_file_names)
        analysis_files=map(lambda x: AsciiDataTable(x,**read_schema(change_extension(x,new_extension="schema"))),analysis_file_names)
        tone_data=[]
        for analysis_file in analysis_files[:]:
            a1_data=analysis_file[["reA1","imA1"]][tone]
            a1_complex=complex(a1_data[0],a1_data[1])
            b1_data=analysis_file[["reB1","imB1"]][tone]
            b1_complex=complex(b1_data[0],b1_data[1])    
            tone_data.append(b1_complex/a1_complex)
        tone_data=np.array(tone_data)
        frequency_tone=analysis_file["Frequency"][tone]
        s1p_complex_data.append([frequency_tone,np.mean(tone_data)])
    s1p=S1PV1(None,sparameter_complex=s1p_complex_data)
    s1p.save(os.path.join(data_directory,"Gamma_{0}.s1p".format(name)))
```

```python
def create_gamma(top_directory):
    """Creates a series of s1p files for each tone spacing in the top directory. The s1p files are b1/a1 without
    ratioing to a4. """
    # Extracting them from the receiver values, do this tone by tone
    for data_directory in os.listdir(top_directory):
        name=data_directory
        data_directory=os.path.join(top_directory,data_directory)
        file_names=os.listdir(data_directory)
        tones=[i for i in range(11)]
        s1p_complex_data=[]
        for tone in tones:
            analysis_file_names=[]
            for file_name in file_names:
                if re.search("txt",file_name,re.IGNORECASE):
                    if re.search("rec",file_name,re.IGNORECASE):
                        if re.search("Tone_{0}_P".format(tone),file_name,re.IGNORECASE):
                            analysis_file_names.append(file_name)
            # Then we load all the single sweeps
            analysis_file_names=map(lambda x:os.path.join(data_directory,x),analysis_file_names)
            analysis_files=map(lambda x: AsciiDataTable(x,**read_schema(change_extension(x,new_extension="schema"))),analysis_file_names)
            tone_data=[]
            for analysis_file in analysis_files[:]:
                a1_data=analysis_file[["reA1","imA1"]][tone]
                a1_complex=complex(a1_data[0],a1_data[1])
                b1_data=analysis_file[["reB1","imB1"]][tone]
                b1_complex=complex(b1_data[0],b1_data[1])    
                tone_data.append(b1_complex/a1_complex)
            tone_data=np.array(tone_data)
            frequency_tone=analysis_file["Frequency"][tone]
            s1p_complex_data.append([frequency_tone,np.mean(tone_data)])
        s1p=S1PV1(None,sparameter_complex=s1p_complex_data)
        s1p.save(os.path.join(data_directory,"Gamma_{0}.s1p".format(name)))
```

```python
top_directories=[r"C:\Share\AWG_Phase_Calibration\Zero_Phase_10Hz_Short",r"C:\Share\AWG_Phase_Calibration\Zero_Phase_10Hz_Open",
r"C:\Share\AWG_Phase_Calibration\Zero_Phase_10Hz_Load",r"C:\Share\AWG_Phase_Calibration\Zero_Phase_Scope_Gamma"]
for top_directory in top_directories[:]:
    create_gamma(top_directory)
```
