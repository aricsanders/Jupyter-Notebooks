```python
from pyMez import *
```

```python
zva=VNA("GPIB::20")
```

```python
frequency_start=50*10**9
frequency_stop=75*10**9
number_points=251
#data_directory=r"C:\Share\WR15_20180213"
data_directory=r"C:\Share\20180215_WR15_VDI"
```

```python
test_frequency_list=np.linspace(frequency_start,frequency_stop,number_points)
# Note with the frequency converter you can't initialize the sweeps. Set up the zva traces manually
# to have "S11","S21","S12","S22","A1_D1","B1_D1","A2_D1","B2_D1"
```

```python

zva.set_IFBW(10)
zva.set_frequency(frequency_start,frequency_stop,number_points)
s2p=zva.measure_sparameters()
name=auto_name(directory=data_directory,specific_descriptor="Short_Port1",general_descriptor="WR15",extension="s2p")
s2p.path=os.path.join(data_directory,name)
s2p.save()
```

```python
s2p.show()
```

```python
#zva.initialize_w2p()
zva.set_IFBW(10)
zva.set_frequency(frequency_start,frequency_stop,number_points)
w2p=zva.measure_w2p()
name=auto_name(directory=data_directory,specific_descriptor="Short_Port1",general_descriptor="WR15_Wave_Parameters",extension="w2p")
w2p.path=os.path.join(data_directory,name)
w2p.save()
w2p.options["IFBW"]=zva.get_IFBW()
w2p.options["Power"]=zva.get_power()
w2p.options["Extender"]="VNAX530"
w2p.save_schema(change_extension(os.path.join(data_directory,name),new_extension="schema"))
```

```python
# now we change the ones used int he MUF to GHz
directory=r"C:\Share\MUF_WR15\Raw Measurements"
for s2p_file in os.listdir(directory):
    new_s2p=S2PV1(os.path.join(directory,s2p_file))
    new_s2p.change_frequency_units("GHz")
    new_s2p.save()
```

```python
# Now we join the files for the short
short_1=S2PV1(os.path.join(directory,r"Short_Port1_WR15_20180213_002.s2p"))
short_2=S2PV1(os.path.join(directory,r"Short_Port2_WR15_20180213_001.s2p"))
new_data=[[short_1.data[row_index][0],short_1.data[row_index][1],
           short_1.data[row_index][2],short_1.data[row_index][3],short_1.data[row_index][4],
           short_1.data[row_index][5],short_1.data[row_index][6],short_2.data[row_index][7],
          short_2.data[row_index][8]] for row_index,row in enumerate(short_1.data)]
options=short_1.options.copy()
options["data"]=new_data
options["sparameter_complex"]=[]
combined_short=S2PV1(None,**options)
combined_short.path=os.path.join(directory,r"Combined_Short_WR15_20180213_001.s2p")
combined_short.save()
```

```python
short_1.show();
short_2.show()
combined_short.show();
```
