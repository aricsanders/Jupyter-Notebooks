```python
import pyMeasure.Code.DataHandlers.GeneralModels as GeneralModels
import os
os.chdir(GeneralModels.TESTS_DIRECTORY)
file_name="ftest6_L1_g5_HF_air"
options={"header_begin_line":0,"header_end_line":2,"column_names_begin_line":2,
         "column_names_end_line":3,"data_begin_line":4,"data_end_line":None,"column_names_delimiter":' ',
        "column_names_begin_token":'#',"column_names_end_token":'\n',"data_table_element_separator":None,
         "data_delimiter":' ',"comment_begin":"#",
         "comment_end":"\n","row_end_token":'\n',"column_types":['float' for i in range(9)],

        }
rfs=""
for i in range(9):
    if i==8:
        rfs=rfs+"{%s:.6g}"%(str(i))
    else:
        rfs=rfs+"{%s:.6g}{delimiter}"%(str(i))
options["row_formatter_string"]=rfs
data_table=GeneralModels.AsciiDataTable(file_name,**options)
print data_table
```
