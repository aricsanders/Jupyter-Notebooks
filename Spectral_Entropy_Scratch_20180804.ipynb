```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_vector=np.random.normal(loc=0,scale=.03,size=1024)
random_vector=np.sin([i for i in range(1024)])
random_vector=np.random.uniform(-1,1,size=1024)
```

```python
plt.plot(random_vector)
plt.show()
```

```python
# now we can find the fft
fft_random=np.fft.fft(random_vector)
```

```python
plt.plot(fft_random.real)
plt.show()
```

```python
# power spectrum
plt.plot(fft_random*np.conjugate(fft_random))
plt.show()
```

```python
# auto correlation
plt.plot(np.fft.ifft(fft_random*np.conjugate(fft_random)))
plt.show()
```

```python
power_spectrum=fft_random*np.conjugate(fft_random)
probability=1/sum(power_spectrum)*power_spectrum
plt.plot(probability)
plt.show()
```

```python
entropy=-sum(abs(probability)*np.log2(abs(probability)))/np.log2(len(probability))
```
