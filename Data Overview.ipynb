So now what if we want to write a function, f such that f maps Dylan/Rich data to XML

$f:Dylan->XML$

```python
def f(file_path=r"F:\Share\CTL Computing\Data For Playing\thru.s2p"):
    try:
        test_file=numpy.loadtxt(file_path)
    except:
        print("Numpy has failed to read the file properly")
        raise
    column_heads=['Index','Frequency','Re_S11','Im_S11','Re_S12','Im_S12','Re_S21','Im_S21','Re_S22','Im_S22']



    # Now we can loop it to create an imput dictionary
    data_values=[]
    for line_number,line in enumerate(test_file):
        ar=line    
        ls=ar.tolist()
        ls.insert(0,line_number+1)
        dy={column_heads[i]:item for i,item in enumerate(ls)}
        data_values.append(dy)
    test_dictionary={'Data_Description':{'Index':'The order it was measured','Frequency':'Frequency in GHZ','Re_S11':'Real part of S11.','Im_S11':'Imaginary part of S11.',
                                     'Re_S21':'Real part of S21.','Im_S21':'Imaginary part of S21.','Re_S12':'Real part of S11.','Im_S12':'Imaginary part of S12.',
                 'Re_S22':'Real part of S22.','Im_S22':'Imaginary part of S22.'},'Data':""}
    test_dictionary["Data"]=data_values
    xml_version=M.DataTable(**test_dictionary)
    return xml_version

out_file=f(r"F:\Share\CTL Computing\Data For Playing\thru.s2p")
out_file.save()
print out_file.get_header()

    
```
