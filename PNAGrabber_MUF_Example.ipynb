# Data Acquistion and Analysis with PNA Grabber and MUF

This document is  a guide to data acquistion with PNAGrabber.exe and error correction with the MUF for a short-open-load-thru (SOLT) calibration method using a E8631A PNA, and commerical 2.4mm calibration kit.

 <h3><a id="overview" style="color:black">Overview</a> </h3>

The measurement process starts with the cleaning of the connectors of the calibration kit. Cleaning the connectors should remove any particulates in the threads of the connector or on the mating faces of the connector. After the connectors are clean, then the thru, a male-to-male adapter, is connected between port one and a cable attached to port two. The PNA is then setup by using the front panel of the PNA itself. The sweep parameters (sweep start, sweep stop and number of points) and the before measurement settling time are all set on the front pannel. The measurement is continued on a remote computer using PNAGrabber which sets the IF-Bandwidth. The calibration kit is measured starting with the thru, then progressing to the shorts, opens and loads. Each measurement is performed with two calibration standards, one on the male end of the thru and one on the female end of the cable. After the measurement the files generated by PNAGrabber are then analyzed by the MUF, which returns the s-parameters corrected and uncertainities.

 <h3><a id="MUF" style="color:black">Correcting Data With the MUF</a> </h3>

Several standard definition files have been collected at //jake.nist.gov/$MUF, and each calibration standard entered into the MUF will require at least 2 inputs, a definition file and a measurement file.

 <h3><a id="Notes" style="color:black">Notes:</a> </h3>

<ol>
<li> Moving files around is difficult, one issue is that the menu files that the MUF / PNAGrabber uses are just pointers to the data in different formats. This causes problems if the data files are moved. The menu files can be made into archives. </li>
<li> There is a lot of back and forth for the measurement so having the control computer close to the actual setup is handy.</li>
<li> Sweep parameters are set on the instrument, but IFBW is set on PNAGrabber </li>

</ol>
