```python
import datetime
import os
import re
import gc
from pyMeasure.Code.DataHandlers.NISTModels import *
```

```python
TOP_DIRECTORY=r'C:\Share\Ck_Std_raw_ascii'
file_names=os.listdir(TOP_DIRECTORY)
metadata_keys=["System_Id","System_Letter","Connector_Type_Calibration","Connector_Type_Measurement",
              "Measurement_Type","Measurement_Date","Measurement_Time","Program_Used","Program_Revision","Operator",
              "Calibration_Name","Calibration_Date","Port_Used","Number_Connects","Number_Repeats","Nbs",
              "Number_Frequencies","Start_Frequency",
              "Device_Description","Device_Id"]
```

```python
raw_files=[]
one_port_raw_files=[]
two_port_raw_files=[]
two_port_NR_raw_files=[]
power_raw_files=[]
for index,file_name in enumerate(file_names[:]):
    in_file=open(os.path.join(TOP_DIRECTORY,file_name),'r')
    lines=[]
    for line in in_file:
        lines.append(line)
    in_file.close()
    #print index,file_name
    if re.search('1-port',lines[4],re.IGNORECASE):
        one_port_raw_files.append(os.path.join(TOP_DIRECTORY,file_name))
    elif re.search('2-port',lines[4],re.IGNORECASE) and not re.search('2-portNR',lines[4],re.IGNORECASE):
        two_port_raw_files.append(os.path.join(TOP_DIRECTORY,file_name))
    elif re.search('2-portNR',lines[4],re.IGNORECASE):
        two_port_NR_raw_files.append(os.path.join(TOP_DIRECTORY,file_name))
    elif re.search('Thermistor|Dry Cal',lines[4],re.IGNORECASE):
        power_raw_files.append(os.path.join(TOP_DIRECTORY,file_name))

```

```python
print("There are %s one port raw files"%len(one_port_raw_files))
print("There are %s two port raw files"%len(two_port_raw_files))
print("There are %s two port NR raw files"%len(two_port_NR_raw_files))
print("There are %s power raw files"%len(power_raw_files))
```

```python
measurement_dates=[]
two_port_NR_metadata=[]
for file_name in two_port_NR_raw_files:
    table=TwoPortNRRawModel(file_name)
    timestamp=table.metadata["Measurement_Date"]+" "+table.metadata["Measurement_Time"]
    datetime_timestamp=datetime.datetime.strptime(timestamp,'%d %b %Y %H:%M:%S')
    measurement_dates.append(datetime_timestamp.isoformat())
    metadata_keys=["System_Id","System_Letter","Connector_Type_Calibration","Connector_Type_Measurement",
              "Measurement_Type","Measurement_Date","Measurement_Time","Program_Used","Program_Revision","Operator",
              "Calibration_Name","Calibration_Date","Port_Used","Number_Connects","Number_Repeats","Nbs",
              "Number_Frequencies","Start_Frequency",
              "Device_Description","Device_Id"]
    metadata=[table.metadata[key] for key in metadata_keys]
    two_port_NR_metadata.append(metadata)
measurement_table_data=[]
for index,date in enumerate(measurement_dates):
    measurement_table_data.append([date]+two_port_NR_metadata[index])
measurement_table_column_names=['Measurement_Timestamp']+metadata_keys
options={"data":measurement_table_data,"column_names":measurement_table_column_names,
        "column_types":['str' for column in measurement_table_column_names]}
ascii_measurement_table=AsciiDataTable(None,**options)
print ascii_measurement_table.get_column('Measurement_Timestamp')
```

```python
measurement_dates=[]
two_port_metadata=[]
for file_name in two_port_raw_files:
    table=TwoPortRawModel(file_name)
    timestamp=table.metadata["Measurement_Date"]+" "+table.metadata["Measurement_Time"]
    datetime_timestamp=datetime.datetime.strptime(timestamp,'%d %b %Y %H:%M:%S')
    measurement_dates.append(datetime_timestamp.isoformat())
    metadata_keys=["System_Id","System_Letter","Connector_Type_Calibration","Connector_Type_Measurement",
              "Measurement_Type","Measurement_Date","Measurement_Time","Program_Used","Program_Revision","Operator",
              "Calibration_Name","Calibration_Date","Port_Used","Number_Connects","Number_Repeats","Nbs",
              "Number_Frequencies","Start_Frequency",
              "Device_Description","Device_Id"]
    metadata=[table.metadata[key].replace(',','-') for key in metadata_keys]
    two_port_metadata.append(metadata)
measurement_table_data=[]
for index,date in enumerate(measurement_dates):
    measurement_table_data.append([date]+two_port_metadata[index])
measurement_table_column_names=['Measurement_Timestamp']+metadata_keys
options={"data":measurement_table_data,"column_names":measurement_table_column_names,
        "column_types":['str' for column in measurement_table_column_names]}
ascii_measurement_table=AsciiDataTable(None,**options)
print ascii_measurement_table.get_column('Measurement_Timestamp')
```
